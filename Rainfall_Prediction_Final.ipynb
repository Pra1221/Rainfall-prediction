```python
#drop correlated columns
data = data.drop(columns = ['maxtemp','temparature','mintemp'])
```

```python
data.head()
```

```python
print(data['rainfall'].value_counts())
```

```python
#separate majority and minority class
df_majority = data[data['rainfall']==1]
df_minority = data[data['rainfall']==0]
```

```python
print(df_majority.shape)
print(df_minority.shape)
```

```python
#down sampling majority class to match minority count

df_majority_downsampled = resample(df_majority,
                                  replace=False,
                                  n_samples=len(df_minority),
                                  random_state=42)
```

```python
print(df_majority_downsampled.shape)
```

```python
df_downsampled = pd.concat([df_majority_downsampled,df_minority])
```

```python
df_downsampled.shape
```

```python
df_downsampled.head()
```

```python
#shuffle the final dataframe
df_downsampled = df_downsampled.sample(frac=1,random_state=42).reset_index(drop=True)
```

```python
df_downsampled.head()
```

```python
df_downsampled['rainfall'].value_counts()
```

```python
#split feature and target as x and y
X = df_downsampled.drop(columns=['rainfall'])
Y = df_downsampled['rainfall']
```

```python
print(X)
```

```python
print(Y)
```

```python
# splitting the data into training data and test data
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

Model Training

```python
# 2. Define Models
models = {
    "Random Forest": RandomForestClassifier(random_state=42),
    "Logistic Regression": LogisticRegression(max_iter=1000, solver='liblinear'),
    "XGBoost": XGBClassifier(eval_metric='logloss', random_state=42)
}
```

```python
# 3. Hyperparameter Tuning Example for RF
param_grids = {
    "Random Forest": {
        "n_estimators": [100],
        "max_depth": [None, 10],
    },
    "Logistic Regression": {},
    "XGBoost": {
        "n_estimators": [100],
        "max_depth": [3, 5],
    }
}
```

```python
best_model = None
best_score = -1
best_name = ""
results = []

# 4. Train & Evaluate
for name, model in models.items():
    print(f"\nTraining {name}...")
    if param_grids[name]:
        grid = GridSearchCV(model, param_grids[name], cv=5, scoring='f1', n_jobs=-1)
        grid.fit(X_train, y_train)
        model = grid.best_estimator_
    else:
        model.fit(X_train, y_train)

    y_pred = model.predict(X_test)
    score = f1_score(y_test, y_pred)
    print(f"{name} F1 Score: {score:.4f}")

    results.append((name, score))

    if score > best_score:
        best_score = score
        best_model = model
        best_name = name
```

```python
print("\nModel Comparison:")
for name, score in results:
    print(f"{name}: F1 Score = {score:.4f}")
print(f"\n✅ Best Model: {best_name} with F1 Score = {best_score:.4f}")
```
